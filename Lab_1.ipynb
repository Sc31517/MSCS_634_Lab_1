#Step 3: Data Preprocessing

# 1.Handling Missing Values

```python
#Handling Missing Values
# Check missing values
df.isnull().sum()
```

```python
#Handle Missing Values
df.head()
```

```python
#Handle Missing Values
# Fill missing numeric values with mean
df["Sales"].fillna(df["Sales"].mean(), inplace=True)
df["Profit"].fillna(df["Profit"].mean(), inplace=True)

# Fill missing categorical values with mode
df["Category"].fillna(df["Category"].mode()[0], inplace=True)
```

```python
#Dataset AFTER Handling Missing Values
df.isnull().sum()
```

#2.Outlier Detection and Removal (IQR Method)

```python
#Calculate IQR for Sales
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)
IQR = Q3 - Q1

Q1, Q3, IQR
```

```python
#Identify Outliers
outliers = df[(df["Sales"] < (Q1 - 1.5 * IQR)) |
              (df["Sales"] > (Q3 + 1.5 * IQR))]

outliers
```

```python
#Remove Outliers
df_no_outliers = df[(df["Sales"] >= (Q1 - 1.5 * IQR)) &
                    (df["Sales"] <= (Q3 + 1.5 * IQR))]

df_no_outliers.head()
```

#3. Data Reduction

```python
#Dataset Before the Reduction
df_no_outliers.shape
```

```python
#Sampling (50% of data)
df_sampled = df_no_outliers.sample(frac=0.5, random_state=42)
df_sampled.shape
```

```python
#Dimension Elimination (
df_reduced = df_sampled.drop(columns=["Row ID", "Postal Code"])
df_reduced.head()

```

#4. Data Scaling and Discretization

```python
#Dataset BEFORE Scaling
df_reduced[["Sales", "Profit"]].head()

```

```python
#Min-Max Scaling
df_reduced["Sales_Scaled"] = (
    df_reduced["Sales"] - df_reduced["Sales"].min()
) / (
    df_reduced["Sales"].max() - df_reduced["Sales"].min()
)

df_reduced["Profit_Scaled"] = (
    df_reduced["Profit"] - df_reduced["Profit"].min()
) / (
    df_reduced["Profit"].max() - df_reduced["Profit"].min()
)

```

```python
#Discretization (Profit Categories)

df_reduced["Profit_Category"] = pd.cut(
    df_reduced["Profit"],
    bins=[-1000, 0, 200, 1000],
    labels=["Loss", "Low Profit", "High Profit"]
)

df_reduced[["Profit", "Profit_Category"]].head()

```

#2. Central Tendency Measures

```python
#Handled correctly to avoid the mode() error
# Central Tendency Measures for Sales
sales_min = df_reduced["Sales"].min()
sales_max = df_reduced["Sales"].max()
sales_mean = df_reduced["Sales"].mean()
sales_median = df_reduced["Sales"].median()
sales_mode = df_reduced["Sales"].mode()

sales_min, sales_max, sales_mean, sales_median, sales_mode
```

```python
#Clean Table
import pandas as pd

central_tendency = pd.DataFrame({
    "Measure": ["Minimum", "Maximum", "Mean", "Median", "Mode"],
    "Sales Value": [
        sales_min,
        sales_max,
        sales_mean,
        sales_median,
        sales_mode.iloc[0]
    ]
})

central_tendency
```

#3. Dispersion Measures

```python
# Range
sales_range = sales_max - sales_min

# Quartiles
Q1 = df_reduced["Sales"].quantile(0.25)
Q2 = df_reduced["Sales"].quantile(0.50)
Q3 = df_reduced["Sales"].quantile(0.75)

# Interquartile Range
IQR = Q3 - Q1

# Variance and Standard Deviation
variance = df_reduced["Sales"].var()
std_dev = df_reduced["Sales"].std()

sales_range, Q1, Q2, Q3, IQR, variance, std_dev
```
